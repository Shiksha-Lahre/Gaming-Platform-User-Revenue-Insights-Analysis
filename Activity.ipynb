```python
#changing datatype (from object to datetime) of signup and last signin date
df['Last_Login'] = pd.to_datetime(df['Last_Login'])
df['Signup_Date'] = pd.to_datetime(df['Signup_Date'])
```

```python
# Ensuring datetime format for proper comparison
df = df[df['Last_Login'] > df['Signup_Date']]
```

```python
df.info()
```

```python
df['Total_Revenue_USD'].sum()
```

```python
# Creating additional columns for time granularity
df['login_date'] = df['Last_Login'].dt.date
df['login_week'] = df['Last_Login'].dt.to_period('W').apply(lambda r: r.start_time.date())
df['login_month'] = df['Last_Login'].dt.to_period('M').apply(lambda r: r.start_time.date())

# Daily active users
dau = df.groupby('login_date')['User_ID'].nunique().reset_index(name='DAU')

# Week;y active users
wau = df.groupby('login_week')['User_ID'].nunique().reset_index(name='WAU')

# Monthly active users
mau = df.groupby('login_month')['User_ID'].nunique().reset_index(name='MAU')
```

```python
#Firstly, ensuring the 'login_date' is in datetime format

dau['login_date'] = pd.to_datetime(dau['login_date'])
dau['Weekday'] = dau['login_date'].dt.day_name()
```

```python
### Behavioral Pattern: Active days and usage frequency
```

```python
# Most active weekdays
df['Weekday'] = df['Last_Login'].dt.day_name()
active_days = df['Weekday'].value_counts().reset_index()
active_days.columns = ['Weekday', 'Login_Count']
```

```python
# User login frequency
user_login_freq = df.groupby('User_ID')['Last_Login'].nunique().reset_index(name='Login_Days')
```

```python
active_days
```

```python
user_login_freq
```

```python
avg_login_days = user_login_freq['Login_Days'].mean()
print(avg_login_days)
```

```python
# Early Signs of Churn
# 1. Long Gaps Between Sign-up and Last Login

df['Days_To_Churn'] = (df['Last_Login'] - df['Signup_Date']).dt.days
early_churners = df[df['Days_To_Churn'] <= 3]

# 2. Low Session or Playtime Users

low_session_users = df[df['Total_Play_Sessions'] <= 2]
low_time_users = df[df['Total_Hours_Played'] < 1]
```

```python
early_churners
```

```python
low_session_users
```

```python
low_time_users
```

```python
# Count of early churners by country
early_churn_by_country = early_churners['Country'].value_counts().reset_index()
early_churn_by_country.columns = ['Country', 'Early_Churn_Count']

early_churn_by_country
```

```python
# High-Value / High-Retention Users

# Top 10% users by revenue
high_value_threshold = df['Total_Revenue_USD'].quantile(0.90)
high_value_users = df[df['Total_Revenue_USD'] >= high_value_threshold]

high_value_users
```

```python
# Longest retained users (maximum differece between signup and last login)
retained_users = df.sort_values(by='Days_To_Churn', ascending=False).head()
retained_users
```

```python
top_10_revenue = high_value_users['Total_Revenue_USD'].sum()
print("Revenue from top 10% users: $", round(top_10_revenue, 2))

top_10_revenue
```

```python
# Suggestions to Improve Retention & Revenue

if early_churners.shape[0] / df.shape[0] > 0.3:
    print("🚨 High churn in first 3 days: consider onboarding improvements")

if df['In_Game_Purchases_Count'].mean() < 2:
    print("💰 In-game purchases are low — introduce time-limited offers or reward systems")

if df['Preferred_Game_Mode'].value_counts().idxmax() == 'Solo':
    print("🎮 Solo mode is most popular — expand content for solo players")

if df['Device_Type'].value_counts().idxmax() == 'Mobile':
    print("📱 Mobile dominates — optimize interface and ad monetization for mobile users")
```
